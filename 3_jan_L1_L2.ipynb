```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# independent
x = np.array([-1,0,1,2,3,4])
# dependent
y = np.array([-3,-1,1,3,5,7])
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel('Independent')
plt.ylabel('Dependent')
plt.show()
```

```python
x.shape
```

```python
x = x.reshape(-1,1)
```

```python
x
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='red')
plt.xlabel('Independent')
plt.ylabel('Dependent')
plt.show()
```

```python
# lasso & ridge Regression
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import Ridge
```

```python
rmodel = Ridge(alpha=5.0)
```

```python
rmodel.fit(x,y)
```

```python
rmodel.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,rmodel.predict(x),color='red')
plt.xlabel('Independent')
plt.ylabel('Dependent')
plt.show()
```

```python
test = np.array([[10.0],[20.0]])
```

```python
rmodel.predict(test)
```
