```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
dataset = pd.read_csv('ad.csv',index_col='Unnamed: 0')
```

```python
dataset
```

```python
x1 = dataset.iloc[:,0:3]
```

```python
x1
```

```python
y = dataset.iloc[:,3]
```

```python
y
```

```python
import statsmodels.api as sm
```

```python
x_constant = sm.add_constant(x1)
```

```python
x_constant
```

```python
lin_reg = sm.OLS(y,x_constant).fit()
```

```python
lin_reg.summary()
```

```python
#Autocorrelation
```

```python
#DW test score ==> 2.08
#means we have little Autocorelation
```

```python
#rainBow
```

```python
sm.stats.diagnostic.linear_rainbow(res=lin_reg)
```

```python
# second value is p-value
# h0 = your data is linear
# 0.71 > 0.05
# we accept the Null Hypo
```

```python
sns.regplot(x=lin_reg.predict(x_constant),y=lin_reg.resid)
```

```python
# Homosecadastcity
```

```python
import statsmodels.stats.api as sms
```

```python
sms.het_goldfeldquandt(lin_reg.resid,lin_reg.model.exog)
```

```python
# second value is p-value
# h0 = your data is homoscedasticity
# 0.07 > 0.05
# we accept the hypoth
```
